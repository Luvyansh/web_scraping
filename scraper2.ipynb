```python
from bs4 import BeautifulSoup
import glob
import pandas as pd

# Initialize an empty list to store the data
data = []

# Loop through all HTML files matching the pattern
for filename in glob.glob('biz-*.html'):
    with open(filename, 'r', encoding='utf-8') as file:
        soup = BeautifulSoup(file, 'html.parser')
        
        # Extract the relevant information
        for business in soup.find_all('div', class_='business'):
            business_id = business.get('data-business-id')
            latitude = business.get('data-latitude')
            longitude = business.get('data-longitude')
            
            if business_id and latitude and longitude:
                data.append({
                    'business_id': business_id,
                    'latitude': float(latitude),
                    'longitude': float(longitude)
                })

# Convert to DataFrame
df_businesses = pd.DataFrame(data)
print(df_businesses.head())
```
