```python
#Write a Python script using requests and BeautifulSoup to extract all the headings (h2) from a given URL.
import requests
from bs4 import BeautifulSoup as bs
import time

url = "https://books.toscrape.com/index.html"
r = requests.get(url).text
soup = BeautifulSoup(r, "html.parser")

h2 = soup.find_all("h1")
for i in h2:
    print(i.text + '\n')
```

```python
#Given a list of product prices as strings (e.g., ["₹1,200", "₹999", "₹2,499"]), write code to extract the numeric values as integers.
import re

arr = ["₹1,200", "₹999", "₹2,499"]
for i in arr:
    number = int(re.sub(r"[^\d]", "", i))
    print(number)
```

```python
#Use a loop to scrape book titles from the first 2 pages of https://books.toscrape.com and print them.
for page in range(1, 3):
    url = f'https://books.toscrape.com/catalogue/page-{page}.html'
    r = requests.get(url).text
    soup = bs(r, 'html.parser')
    titles = soup.find_all('a', title=True)
    for title in titles:
        value = title.get_attribute_list('title')
        print(re.sub(r"[^\w ]", "", str(value)))
```

```python
#Write a script that fetches and prints all the image URLs (img src="") from a webpage using BeautifulSoup.
url = 'https://books.toscrape.com/'
r = requests.get(url).text
soup = bs(r, 'html.parser')
imgs = soup.find_all('img', alt=True)
for img in imgs:
    print(url + re.sub(r"[\'\[\]]", "", str(img.get_attribute_list('src'))))
```

```python
#Create a dictionary from a webpage where keys are quote texts and values are authors using requests and bs4. (You can use quotes.toscrape.com)
url = "https://quotes.toscrape.com/page/{}/"
cur_url = url
page = 1

while True:
    time.sleep(1) #reducing server load
    r = requests.get(cur_url.format(page)).text
    quotes = dict()
    soup = bs(r, 'html.parser')
    divs = soup.find_all('div', class_='quote')
    for div in divs:
        text = div.find('span', class_='text').get_text()
        author = div.find('small', class_='author').get_text()
        quotes[text] = author
    
    for quote in quotes:
        print(quote + " : " + quotes[quote])
    
    nex = soup.find('li', class_='next')
    if not nex:
        break
    page += 1
```

```python
#Write a function that accepts a URL and returns the total number of links (a tags) on that page
def noa(url):
    r = requests.get(url).text
    soup = bs(r, 'html.parser')
    count = 1
    a = soup.find_all('a')
    for i in a:
        count += 1
    return count

url = 'https://quotes.toscrape.com/page/1/'
noa(url)
```

```python
#Extract all list items from a webpage's ul or ol tag and print their text content.
url = 'https://books.toscrape.com/index.html'
r = requests.get(url)
r.encoding = 'utf-8'
soup = bs(r.text, 'html.parser')
uls = soup.find_all('ul')
ols = soup.find_all('ol')
for ul in uls:
    lis = ul.find_all('li')
    for li in lis:
        print(li.get_text(strip = True))
for ol in ols:
    lis = ol.find_all('li')
    for li in lis:
        print(li.get_text(strip = True))
```

```python
#Write Python code to simulate a login form using requests.Session() and scrape data after logging in. (They may give you quotes.toscrape.com/login as example.)
url = 'https://quotes.toscrape.com/login'
base_url = 'https://quotes.toscrape.com/page/{}/'
session = requests.Session()
login_page = session.get(url).text
soup = bs(login_page, 'html.parser')
csrf_token = soup.find('input', {"name": "csrf_token"}).get('value')
payload = {
    "csrf_token": csrf_token,
    "username": "admin",
    "password": "admin"
}
res = session.post(url, data=payload)
if "Logout" not in res.text:
    print("Login Failed !")
else:
    print("Logged In Successfully !!!")
page = 1
while True:
    cur_url = base_url.format(page)
    time.sleep(1)
    r = session.get(cur_url).text
    soup = bs(r, 'html.parser')
    print(f"Page {page} Quotes:")
    qts = soup.find_all("span", class_="text")
    for qt in qts:
        print(qt.get_text())
    page += 1
    nex = soup.find('li', class_="next")
    if not nex:
        break
    else:
        continue
```
