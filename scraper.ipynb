```python
import sqlite3
import pandas as pd

# Connect to the SQLite database
conn = sqlite3.connect('violations.db')

# Query to count Moderate Risk violations on Fridays
query = """
SELECT business_id, COUNT(*) as violation_count
FROM violations
WHERE risk_category = 'Moderate Risk' AND strftime('%w', date) = '5'
GROUP BY business_id
"""
moderate_risk_violations = pd.read_sql_query(query, conn)

conn.close()
```

```python
from bs4 import BeautifulSoup
import glob

# Function to extract business_id and postal_code from an HTML file
def extract_business_data(file):
    with open(file, 'r') as f:
        soup = BeautifulSoup(f, 'html.parser')
        # Assuming the structure contains tags with business_id and postal_code
        business_id_tag = soup.find('span', {'class': 'business_id'})
        postal_code_tag = soup.find('span', {'class': 'postal_code'})
        
        # Check if the tags were found
        business_id = business_id_tag.text if business_id_tag else None
        postal_code = postal_code_tag.text if postal_code_tag else None
        
        if not business_id or not postal_code:
            print(f"Error in file: {file}")
            print(f"business_id: {business_id_tag}")
            print(f"postal_code: {postal_code_tag}")
        
        return business_id, postal_code

# List to store extracted data
business_data = []

# Loop through all HTML files
for file in glob.glob('biz-*.html'):
    business_id, postal_code = extract_business_data(file)
    if business_id and postal_code:  # Only append if both are found
        business_data.append((business_id, postal_code))

# Convert the extracted data to a DataFrame
business_df = pd.DataFrame(business_data, columns=['business_id', 'postal_code'])

# Filter businesses with postal code 94104
filtered_businesses = business_df[business_df['postal_code'] == '94104']

# Merge with the violation counts
result = pd.merge(filtered_businesses, moderate_risk_violations, on='business_id', how='left')

# Sum the violation counts for businesses in postal code 94104
total_moderate_risk_violations = result['violation_count'].sum()

# Handle cases where there are no violations
if pd.isna(total_moderate_risk_violations):
    total_moderate_risk_violations = 0

print(f"Total number of moderate risk violations for postal code 94104: {total_moderate_risk_violations}")
```
